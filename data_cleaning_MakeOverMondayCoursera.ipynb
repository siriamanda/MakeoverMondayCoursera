```python
# Import the libraries

import pandas as pd
import numpy as np

import pycountry_convert as pc
```

```python
# Import the data

data = pd.read_excel('World Indicators.xlsx')
```

```python
data.head()
```

```python
# Create list with country codes

country_code = []

for i in data['Country']:
    code = pc.country_name_to_country_alpha2(i, cn_name_format = "default")
    country_code.append(code)
```

```python
# Create list with continents

continents = []
for i in country_code:
    continent = pc.country_alpha2_to_continent_code(i)
    continents.append(continent)
```

```python
# Add new column with the continent code

data['Continent'] = continents
```

```python
data
```

```python
data_1 = data[['Country', 'Continent', 'Year', 'Life expectancy at birth, female (years)', 'Life expectancy at birth, male (years)']]
```

```python
data_1.to_csv('cleaned_1.csv')
```

```python
# Create dataframe with African countries

data_AF = data_1.loc[data_1['Continent'] == 'AF']
```

```python
data_AF
```

```python
# Reset index

data_AF.reset_index(drop = True, inplace = True)
```

```python
data_AF
```

```python
# Check datatypes

data_AF.dtypes
```

```python
# Drop null values

data_AF = data_AF.dropna(subset=['Life expectancy at birth, female (years)', 'Life expectancy at birth, male (years)'])
```

```python
# Reset index

data_AF.reset_index(drop = True, inplace = True)
```

```python
data_AF
```

```python
data_AF['Average Life Expectancy'] = (data_AF['Life expectancy at birth, female (years)'] + data_AF['Life expectancy at birth, male (years)']) / 2
```
